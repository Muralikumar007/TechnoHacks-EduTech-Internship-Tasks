## Feature Transformation

```python
df['sex'] = df['sex'].apply({'male':0, 'female':1}.get)
df['smoker'] = df['smoker'].apply({'yes':1, 'no':0}.get)
df['region'] = df['region'].apply({'southwest':1, 'southeast':2, 'northwest':3, 'northeast':4}.get)
```

```python
df.head()
```

## Split the Train and Test 

```python
X = df.drop(['charges', 'sex'], axis=1)
y = df.charges
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
print("y_train shpae: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

## Modeling

### Linear Regression Model

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, y_train)
linreg_pred = linreg.predict(X_test)
```

```python
print("R2 score: ",(r2_score(y_test, linreg_pred)))
```

```python
plt.scatter(y_test, linreg_pred)
plt.xlabel('Y test')
plt.ylabel('Y pred')
plt.show()
```

### Insurance cost of New Customer Linear Regression Model

```python
data = {'age':50, 'bmi':25, 'children':2, 'smoker':1, 'region':2}
index = [0]
cust_df = pd.DataFrame(data, index)
cust_df
```

```python
cost_pred = linreg.predict(cust_df)
print("The medical insurance cost of the new customer is: ", cost_pred)
```

### Random Forest Regression

```python
rfreg = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rfreg.fit(X_train, y_train)
rfreg_pred = rfreg.predict(X_test)
```

```python
print("R2 score: ",(r2_score(y_test, rfreg_pred)))
```

```python
plt.scatter(y_test, rfreg_pred)
plt.xlabel('Y test')
plt.ylabel('Y pred')
plt.show()
```

```python
data = {'age':50, 'bmi':25, 'children':2, 'smoker':1, 'region':2}
index = [0]
cust_df = pd.DataFrame(data, index)
cust_df
```
